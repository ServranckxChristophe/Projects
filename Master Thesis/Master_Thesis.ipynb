### Partie I : SED SIMBAD - VIZIER

Définissons d'abord une fonction pour la SED & qui aura pour rôle de convertir les fréquences en longueur d'onde et d'exprimer le flux dans les bonnes unités.

```python
def conversion_Jy_to_Watt(SED):
    """
    1) Cette fonction lit un fichier SED contenant les données utiles à la réalisation
    des SED & extrait les fréquences et flux ;
    2) Les fréquences sont converties en longueurs d'onde ; et
    3) Les flux sont exprimés dans les unités adéquates.
    """
    
    #On initialise nos listes (de taille indéfinie) pour l'extraction des données
    freq = []
    fluxJan = []

    #On ouvre le fichier et on extrait les colonnes dans les vecteurs initialisés
    with open(SED, 'r', encoding="utf-8") as datafile: #permet d'ouvrir sans risque que le fichier reste ouvert
        for i in (datafile):
            freq.append(float((i.split()[2])))   #On ajoute à c1 élément par élément
            fluxJan.append(float((i.split()[3])))   #On ajoute à c2 élément par élément

    #On ordonne par ordre croissant les fréquences et les flux correspondants        
    freq, fluxJan = zip(*sorted(zip(freq, fluxJan)))
    
    #On convertit les fréquences en longueur d'onde
    lamb= c/(np.copy(freq)*10**9) * 10**6                    #En µm

    #On change les unités
    flux= np.copy(fluxJan)*10**(-26)*np.copy(freq)*10**9     #En W/m²
    
    return (lamb,flux)
```

### Partie II : Synthèse spectrale
---

Définissons une fonction pour l'extraction de nos données et la conversion des unités.

```python
def conversion_erg_to_watt(SEDP1,SEDP2):
    """
    1) Cette fonction lit un fichier SED contenant les données utiles à la réalisation
    des SED & extrait les fréquences et flux ;
    2) Les fréquences sont converties en longueurs d'onde ; et
    3) Les flux sont exprimés dans les unités adéquates.
    """
    
    #On initialise nos listes (de taille indéfinie) pour l'extraction des données
    lamb1 = []
    flux1 = []
    lamb2 = []
    flux2 = []

    #On ouvre le fichier et on extrait les colonnes dans les vecteurs initialisés
    with open(SEDP1, 'r', encoding="utf-8") as datafile1: #permet d'ouvrir sans risque que le fichier reste ouvert
        for i in (datafile1):
            lamb1.append(float((i.split()[0])))   #On ajoute à c1 élément par élément
            flux1.append(float((i.split()[2])))   #On ajoute à c2 élément par élément

    #On ouvre le fichier et on extrait les colonnes dans les vecteurs initialisés
    with open(SEDP2, 'r', encoding="utf-8") as datafile2: #permet d'ouvrir sans risque que le fichier reste ouvert
        for i in (datafile2): 
            lamb2.append(float((i.split()[0])))   #On ajoute à c1 élément par élément
            flux2.append(float((i.split()[2])))   #On ajoute à c2 élément par élément
            
    #On fusionne les listes
    lam = np.concatenate((lamb1, lamb2))                #En Å
    lF_lbda = np.concatenate((flux1, flux2))            #En erg/cm²/s
    
    #On ordonne les deux listes en préservant les correspondances
    lam, lF_lbda = zip(*sorted(zip(lam, lF_lbda)))

    #On change lambda d'unité pour l'axe x
    lamb = np.copy(lam) * 10**(-4)                      #En µm
    
    F_lamb = np.copy(lF_lbda) / np.copy(lam)            #En erg/cm²/s/Å
    
    #Conversion 
    nu = c/(lamb*10**(-6))                              #En Hz
    nuF_nu = F_lamb* 10**(-3) * (lamb*10**(-6)/nu) * nu #En W/m²/s/Hz
    return (lamb, nuF_nu)
```

Maintenant que notre fonction est réalisée, nous pouvons réaliser le graphique de la SED synthétisée superposée à la SED observée.

```python
lamb_synth,flux_synth = conversion_erg_to_watt('s2600_g-0.5_m1.0_t02_x3_z 0.00_a 0.75_c 1.00_n 0.00_o 0.75_r 0.00_s 0.00.mod_3800.-8900_', 's2600_g-0.5_m1.0_t02_x3_z 0.00_a 0.75_c 1.00_n 0.00_o 0.75_r 0.00_s 0.00.mod_8900.-30000_')

figure = pyplot.figure(figsize = (30,25))  #Configuration de la taille de la figure

#############    SED Observée    #############

pyplot.scatter(lamb_red, flux_red, c='red', s=200, alpha=1, marker='x',label="SED Observée")

###########    SED synthétique    ############

pyplot.plot(lamb_synth, flux_synth, c='blue', alpha=0.5, label="SED synthétique $\star$ carbonée") 

pyplot.title("\n Distribution d'Énergie Spectrale (SED) de V Hya \n", fontsize = 55)
pyplot.ylabel('$\\nu$F($\\nu$) (W m$^{-2}$)\n',fontsize=60)
pyplot.xlabel('\n $\lambda$ ($\mu$m)',fontsize=60)
pyplot.ylim(10**(-16), 10**(-7))
pyplot.yscale('log')  #échelle logarithmique
pyplot.xscale('log')  #échelle logarithmique
plt.legend(borderaxespad=0.5,fontsize=40)
pyplot.tick_params(axis='both', which = 'minor', width = 2, length = 5, color = 'black', labelsize=42)
pyplot.tick_params(axis='both', width = 2, length = 10, color = 'black', labelsize=42)
```

##### Extrapolation de la SED synthétique
___

On importe les fichiers/donnés des SED (synthétique & observée).

```python
#Import des fichiers/données et conversion en les bonnes unités

lsynthCarbo,flsynthCarbo = conversion_erg_to_watt('s2600_g-0.5_m1.0_t02_x3_z+0.00_a+0.75_c+1.00_n+0.00_o+0.75_r+0.00_s+0.00.mod_3800.-8900.del0.1', 's2600_g-0.5_m1.0_t02_x3_z+0.00_a+0.75_c+1.00_n+0.00_o+0.75_r+0.00_s+0.00.mod_8900.-30000.del0.1')
lamb_red,flux_red = conversion_Jy_to_Watt('SED_reduit.txt')
```

Commençons par extrapoler la SED synthétique. Pour se faire on regarde le régime dans lequel on se situe. 

```python
#Intervalle de longueur d'onde sur lequel l'extrapolation est faite
lamb_start = 3      #En µm
lamb_end = 1000     #En µm

#On génère les longueurs d'ondes et fréquences de l'intervalle sur lequel on extrapole
x = np.linspace(lamb_start, lamb_end, lamb_end-lamb_start+1)  #En µm                                              #En W.s = W/Hz
nu = c/(x*10**(-6))                                           #En Hz

Teff = 2600            #En K 
```

Nous serons dans le régime de :
* $\underline{\text{Rayleigh-Jeans}}$ si $\mbox{ } hc \ll \lambda k_B T$ (ou $h \nu \ll k T$)
* $\underline{\text{Wien}}$ si $\mbox{ } hc \gg \lambda k_B T$ (ou $h \nu \gg k T$)

```python
if (h*c < lsynthCarbo[-1]*kB*Teff):
    print('On est dans le régime des basses fréquences (grandes longueurs d\'ondes), i.e. \ndans le régime de Rayleigh-Jeans.')
else :
    print('On est dans le régime des grandes fréquences (petites longueurs d\'ondes), i.e. \ndans le régime de Wien.')
```

Nous somme donc dans le régime des basses fréquences et par conséquent nous avons la formule de Rayleigh-Jeans : $B_\lambda(T) = \frac{2 c k_B T}{\lambda^4}$ ou encore en terme de fréquence : $B_\nu(T) = \frac{2 k_B T}{\lambda^2}$.

Il nous reste à tenir compte de la décroissance du flux en l'inverse du carré de la distance à V Hya par la relation : $F = \pi B \left( \frac{R}{d} \right)^2$ , avec :

R, est le rayon de V Hya

d, est la distance à V Hya

Implémentons maintenant ceci sous forme de deux fonctions.

```python
def Planck_BB(Teff,x):
    """
    Réalise la courbe d'un BB pour une température effective, Teff (en K), et des
    longueurs d'onde (en µm) données.
    """
    
    return 2*kB*Teff/(x*10**(-6))**2            #Intensité du BB en W/Hz/m²
```

```python
def Corr_dist_densTOflux(dist,rayon,inten,nu):
    """
    Cette fonction corrige l'intensité (en terme de fréquence) pour la distance 
    séparant la source à l'observateur. Il en résulte une densité de flux qui est 
    elle-même convertie en flux exprimer ici en w/m²
    
    dist : distance à l'étoile reçue en pc
    rayon : rayon de l'étoile reçu en cm
    inten : intensité du BB en W/Hz/m²
    """
    
    d = dist * pc                             #Distance à l'étoile en m
    R = rayon*10**(-2) * (dist/500)           #Rayon de l'étoile en m

    Fnu = inten * np.pi * (R/d)**2            #Densité de flux

    return Fnu * nu * np.pi                   #Flux en W/m²
```

```python
#Prise en compte de la distance
d = 478.51                             #En pc
R = 3.8 *10**(13)                      #Rayon de V Hya en cm

B_nu = Planck_BB(Teff,x)
nuFnu = Corr_dist_densTOflux(d,R,B_nu,nu)

lamb_synth = np.concatenate((lsynthCarbo, x))           #lambda en µm
flux_synth = np.concatenate((flsynthCarbo, nuFnu))       #lambda.flux en erg/cm²/s
```

On trace maintenant notre SED et son extrapolation ainsi que la SED observée.

```python
figure = pyplot.figure(figsize = (30,25))  #Configuration de la taille de la figure

#############   SED Observée    #############

pyplot.scatter(lamb_red, flux_red, c='red', s=200, lw=4, alpha=1, marker='x',label="SED Observée") 

###########   SED synthétique    ###########

pyplot.plot(lamb_synth, flux_synth, c='blue', alpha=0.3, label="SED synthétique $\star$ carbonée") 

#Paramètres du graphique
pyplot.ylabel('$\\nu$F($\\nu$) (W m$^{-2}$)\n',fontsize=60)
pyplot.xlabel('\n $\lambda$ ($\mu$m)',fontsize=60)
pyplot.ylim(10**(-16), 10**(-6))
pyplot.yscale('log')  #échelle logarithmique
pyplot.xscale('log')  #échelle logarithmique nous
#pyplot.title("\n Distribution d'Énergie Spectrale (SED) de V Hya \n", fontsize = 55)
plt.legend(borderaxespad=0.5,fontsize=40)
pyplot.tick_params(axis='both', which = 'minor', width = 2, length = 5, color = 'black', labelsize=42)
pyplot.tick_params(axis='both', width = 2, length = 10, color = 'black', labelsize=42)
```

##### Spectre modèle MARCS

```python
def MARCS_conversion_unit(SEDP1,SEDP2):
    """
    1) Cette fonction lit un fichier SED contenant les données utiles à la réalisation
    des SED & extrait les longueurs d'onde et flux ;
    2) Les longueurs d'onde sont mises en µm au lieu de l'Å; et
    3) Les flux sont exprimés dans les unités adéquates.
    """
    
    #On initialise nos listes (de taille indéfinie) pour l'extraction des données
    lam = []
    lambF_lambda = []

    #On ouvre le fichier et on extrait les colonnes dans les vecteurs initialisés
    with open(SEDP1, 'r', encoding="utf-8") as datafile1: #permet d'ouvrir sans risque que le fichier reste ouvert
        for i in (datafile1):
            lam.append(float((i.split()[0])))   #On ajoute à c1 élément par élément

    #On ouvre le fichier et on extrait les colonnes dans les vecteurs initialisés
    with open(SEDP2, 'r', encoding="utf-8") as datafile2: #permet d'ouvrir sans risque que le fichier reste ouvert
        for i in (datafile2): 
            lambF_lambda.append(float((i.split()[0])))   #On ajoute à c2 élément par élément
    
    #On change lambda d'unité pour l'axe x
    lamb = np.copy(lam) * 10**(-4)                      #En µm
    
    #On calcul la densité de flux
    F_lamb = np.copy(lambF_lambda) / np.copy(lam)       #En erg/s/cm²/Å
    
    #Conversion 
    nu = c/(lamb*10**(-6))                              #En Hz
    nuF_nu = F_lamb* 10**(-3) * (lamb*10**(-6)/nu) * nu #En W/m²/s/Hz
    
    return (lamb, nuF_nu, nuF_nu/nu)
```

Import des fichiers/donné et conversion en les bonnes unités

### Partie III : Corps noir, bande passante & température des grains
---


##### Ajustement d'un "corps noir" pour les données observées
---

La fonction de Planck pour le corps noir est donnée en terme de longueur d'onde par la formule : $B_\lambda(T) = \frac{2 h c^2}{\lambda^5} \frac{1}{e^{hc/ \lambda k T}-1}$, ou en terme de fréquence par : $B_\nu(T) = \frac{2 h \nu^3}{c^2} \frac{1}{e^{h \nu/ k T}-1}$.

Implémentons cette dernière relation pour notre cas.

```python
#On génère une gamme de longueur d'onde comprise entre 1000 et 3800 Å
lbd= np.linspace(1000,3800,3800-1000+1)    #En Å
Int1= lbd *10**(-4)                        #En µm

#On fusionne l'intervalle de longueur d'onde généré avec celui existant
lamb_synth2 = np.concatenate((Int1, lamb_synth))       #En µm
Int = np.concatenate((lamb_synth2,Int3))               #En µm
```

On a maintenant notre intervalle de longueur allant de 1000 à 30000 Å.

```python
#On calcul les fréquences correspondantes
fInt = c/(Int*10**(-6))

Teff = 2600

#On calcul la densité de flux via la formule de Planck
Bnu = (2*h*fInt**3/c**2) * (1/(np.exp((h*fInt)/(kB*Teff))-1))
```

On prend en compte la distance à laquelle se situe V Hya.

```python
nFnu = Corr_dist_densTOflux(d,R,Bnu,fInt)
```

```python
def passband(file):
    """
    Cette fonction extrait les longueurs d'onde et les Réponses Spectrales 
    Relatives (RSR) pour une certaine bande.
    """
    
    #On initialise nos listes (de taille indéfinie) pour l'extraction des données
    lambd = []
    RSR = []

    #On ouvre le fichier et on extrait les colonnes dans les vecteurs initialisés
    with open(file, 'r', encoding="utf-8") as datafile: #permet d'ouvrir sans risque que le fichier reste ouvert
        for i in (datafile):
            lambd.append(float((i.split()[0])))   #On ajoute à c1 élément par élément
            RSR.append(float((i.split()[1])))   #On ajoute à c2 élément par élément

    return (lambd,RSR)
```

##### Intégration numérique - Bande passante
---

On souhaite calculer : $\mbox{ } \nu_{\text{eff}}$F$_\nu = \frac{\int \nu \text{F}_\nu \text{S}_\nu \text{d}\nu}{\int \text{S}_\nu \text{d}\nu}$ ,

où $\text{S}_\nu$ est la bande passante. Il nous faut donc calculer les deux intégrales. Effectuons ceci via la méthode des trapèzes.

Commençons par définir la fonction dont on souhaite calculer l'intégrale.

```python
def BPnuEff(nuFnu,Snu):
    return nuFnu * Snu
```

Définissons la fonction pour calculer l'intégrale via la règle du trapézoïde qui consiste principalement à subdiviser l'aire sous la courbe, tracée par notre fonction, en petits trapèzes...

```python
def int_trap_num(n,SnuJHK,nuFnuJHK,a,b):
    
    Dx=(b-a)/(n-1)  #On définit un intervalle
    In=0
    
    for i in range (n): 
        In = In + BPnuEff(nuFnuJHK[i],SnuJHK[i])
    
    return (In + (-BPnuEff(nuFnuJHK[0],SnuJHK[0])-BPnuEff(nuFnuJHK[-1],SnuJHK[-1]))/2)*Dx
```

```python
def int_trap_denom(n,SnuJHK,a,b):
    
    Dx=(b-a)/(n-1)  #On définit un intervalle
    In=0
    
    for i in range (n): 
        In = In + SnuJHK[i]
    
    return (In + (-SnuJHK[0]-SnuJHK[-1])/2)*Dx
```

Convertissons les longueurs d'onde générée précédements pour l'interpolation en fréquences pour effectuer l'intégration.

```python
nuJ = c/(np.copy(ldaJ)*10**(-6))       #En Hz
nuH = c/(np.copy(ldaH)*10**(-6))       #En Hz
nuK = c/(np.copy(ldaK)*10**(-6))       #En Hz
```

On rassemble toutes nos données dans des vecteurs uniques de même taille pour pouvoir lancer une boucle sur chacune des bandes.

```python
nuJHK = np.empty((3))
lJHK = np.empty((3))
nuFnuJHK = np.empty((3))
SnuJHK = np.empty((3))

lJHK = np.array([ldaJ,ldaH,ldaK])
nuJHK = np.array([nuJ,nuH,nuK])
nuFnuJHK = np.array([ysynJ,ysynH,ysynK])
SnuJHK = np.array([yJ,yH,yK])
```

Nous pouvons maintenant calculer les intégrales et leur rapport pour déterminer $\nu_\text{eff}$F$_\nu$($\nu_\text{eff}$). On définit pour cela les bornes de l'intégrale et le nombre de points (nombre d'intervalle = nombre de points-1) tout en effectuant une boucle pour calculer les intégrales et rapports sur les différentes bandes.

```python
def BandeJHK_nueff(nJHK,SnuJHK,nuFnuJHK):
    
    nuFeff = []
    for i in range(0,3,1):
    
        a = max(nuJHK[i])    #Longueur d'onde min = fréquence max 
        b = min(nuJHK[i])    #Longueur d'onde max = fréquence min 
    
        IntNum = int_trap_num(nJHK[i],SnuJHK[i],nuFnuJHK[i],a,b)
        IntDenom = int_trap_denom(nJHK[i],SnuJHK[i],a,b)
    
        nuFeff.append(IntNum/IntDenom)
    
    return nuFeff
```

```python
nuFeffMARCS = BandeJHK_nueff(nJHK,SnuJHK,nuFnuJHK)
```

Calculons maintenant la longueur d'onde effective. Pour se faire on doit calculer :
$\mbox{ } \lambda_{\text{eff}} = \frac{\int \lambda \text{S}_\lambda \text{d}\lambda}{\int \text{S}_\lambda \text{d}\lambda}$ , où $\text{S}_\lambda$ est la bande passante. 

Il nous faut donc calculer les deux intégrales. Effectuons à nouveau ceci via la méthode des trapèzes. On redéfinit quelque peut nos fonctions

```python
def BPldaEff(lbda,Slbda):
    return lbda * Slbda

def int_trap_num_lbdaEff(n,lbda,SlbdaJHK,a,b):
    
    Dx=(b-a)/(n-1)  #On définit un intervalle
    In=0
    
    for i in range(n): 
        In = In + BPldaEff(lbda[i],SlbdaJHK[i])
    
    return (In + (-BPldaEff(lbda[0],SlbdaJHK[0])-BPldaEff(lbda[-1],SlbdaJHK[-1]))/2)*Dx

def int_trap_denom_lbdaEff(n,SlbdaJHK,a,b):
    
    Dx=(b-a)/(n-1)  #On définit un intervalle
    In=0
    
    for i in range (n): 
        In = In + SlbdaJHK[i]
    
    return (In + (-SlbdaJHK[0]-SlbdaJHK[-1])/2)*Dx
```

Et on calcul finalement $\lambda_{\text{eff}}$ :

```python
def BandesJHK_lbdeff(lJHK,nJHK,SnuJHK):
    
    LambdaEff = []
    for i in range(0,3,1):
    
        a = min(lJHK[i])
        b = max(lJHK[i])
    
        IntNumlbda = int_trap_num_lbdaEff(nJHK[i],lJHK[i],SnuJHK[i],a,b)
        IntDenomlbda = int_trap_denom_lbdaEff(nJHK[i],SnuJHK[i],a,b)
    
        LambdaEff.append(IntNumlbda/IntDenomlbda)
    return LambdaEff
```

##### Intégration du flux bolométrique
---

On souhaite calculer la magnitude bolométrique apparente de V Hya pour voir, à terme, la contribution de son compagnon binaire. Pour cela, nous savons que la magnitude bolométrique absolue par rapport au soleil est donnée par :

$$\text{M}_{\text{bol,}\star} - \text{M}_{\text{bol,}\odot} = -2,5 \log_{10} \left( \frac{\text{L}_{\text{bol,}\star}}{\text{L}_{\text{bol,}\odot}} \right),$$


où L$_{\text{bol,}\odot} = 3,828 \times 10^{26}$ W, la luminosité bolométrique nominale du soleil.

$\text{M}_{\text{bol,}\odot} = 4,74$ mag, la magnitude bolométrique absolue du soleil.


On voit qu'il nous faut donc calculer la luminosité de V Hya. Une autre option est de réexprimer la luminosité en terme du flux bolométrique (flux intégré sur toutes les fréquences) sachant que :

$$\text{L} = \int_\Sigma \text{dA} {\int_0}^\infty {\text{F}_\nu}^+ \text{d}\nu.$$

La première intégrale est égale à :

$$\int_\Sigma \text{dA} = 4 \pi \text{R}^2$$

et représente la surface de l'étoile . La seconde intégrale nous donne quant à elle le flux bolométrique :

$${\int_0}^\infty {\text{F}_\nu}^+ \text{d}\nu = \text{F}_\text{bol}.$$

Cette dernière se calcule facilement au moyen de la méthode des trapèzes. Nous voyons donc qu'à l'aide des équations (13) et (14), l'équation (12) se réexprime comme :

$$\text{L}_\text{bol} = 4 \pi \text{R}^2 \text{F}_\text{bol}$$

En substituant (15) dans (11), on trouve alors :

$$\text{M}_{\text{bol,}\star}  = \text{M}_{\text{bol,}\odot}- 2,5 \log_{10} \left[ \frac{\text{F}_{\text{bol,}\star}}{\text{F}_{\text{bol,}\odot}} \left( \frac{\text{R}_\star}{\text{R}_\odot} \right)^2 \right],$$

avec R$_\star$, le rayon de V Hya donné par l'expression indiquée à l'équation (6) ; et

F$_{\text{bol}, \odot} = 3,226 \times10^{-14}$ W cm$^{-2}$, le flux bolométrique absolu du soleil à l'aide duquel nous trouvons donc la magnitude bolométrique absolue de V Hya : 

```python
def FluxBolo(n,densflux,a,b):
    """
    Cette fonction calcule le flux bolométrique i.e. l'intégral du flux sur 
    toutes les fréquences
    
    n : nombre de points (fréquences)
    densflux : densité du flux en W/m²/Hz
    a : borne inférieure de l'intégrale en Hz
    b : borne supérieure de l'intégrale en Hz
    
    """
    Dx=abs((b-a)/(n-1))  #On définit un intervalle
    In=0

    for i in range (0,n): 
        In = In + densflux[i]
    
    return (In + (-densflux[0]-densflux[-1])/2)*Dx
```

##### Détermination de paramètres du système binaire
---

Déterminons à présent la masse du compagnon de V Hya à l'aide de la fonction de masse :

\begin{equation*}
\text{f}(\text{M}_1,\text{M}_2) = \frac{\text{M}_2^3 \sin^3{\text{i}}}{(\text{M}_1+\text{M}_2)^2} =  \frac{\text{P}}{2 \pi \text{G}} \text{K}_1^3 (1-\text{e}^2)^{3/2}
\end{equation*}

En isolant M$_2$ dans la deuxième égalité on obtient alors l'équation :

\begin{equation*}
    \frac{1}{\text{q}} \text{M}_2^3 - \text{M}_2^2 - 2 \text{M}_1 \text{M}_2 - \text{M}_1^2 = 0 
\end{equation*}

avec q = $\frac{\text{P} \text{K}_1^3}{2 \pi \text{G} \sin^3{i}} (1-\text{e}^2)^{3/2}$. De cette dernière on trouve finalement $\text{M}_2$

```python
z = Symbol('z')

#Paramètres orbitaux
Porb = 9.37*365.25*24*60*60         #En s
K1 = 10.93346 * 10**3     #En m/s
m1 = 1.5 * Msun       #En kg 
e=0.7

q = []

for i in range(len(incl)):
    q.append((Porb*K1**3)/(2 * np.pi * G * np.sin((2*np.pi/360)*incl[i])**3) * (1 - e**2)**(3/2))          #En jours
    sol1, sol2, sol3 = solve((1/q[i]) * z**3 - z**2 - 2 * m1 * z - m1**2)
    #print("Solutions pour un rougissement de %.2f mag :\n" %i)
    #print('\nSolution 1 :', sol1,'kg')
    #print('\nSolution 2 :', sol2,'kg')
    #print('\nSolution 3 :', sol3,'kg \n \n')
    sol3sun = sol1/Msun
    print("La masse du compagnon pour un rougissement de %.2f mag est de %.2f M_sol \n \n" %(i,sol3sun))
```

```python
thetaxs = np.arctan((hAUxs/(2*RintAU))) * (360/(np.pi*2))
inclxs = 90 - thetaxs
print("L'inclinaison du disque pour l'excès IR est de %.2f °" %inclxs)
```

```python
z = Symbol('z')

#Paramètres orbitaux
Porb = 6160*24*60*60         #En s
K1 = 5 * 10**3               #En m/s
m1 = 1.5 * Msun              #En kg 
e = 0.1

qxs = (Porb*K1**3)/(2 * np.pi * G * np.sin((2*np.pi/360)*inclxs)**3) * (1 - e**2)**(3/2)          #En jours
sol1xs, sol2xs, sol3xs = solve((1/qxs) * z**3 - z**2 - 2 * m1 * z - m1**2)
#print("Solutions pour un rougissement de %.2f mag :\n" %i)
#print('\nSolution 1 :', sol1,'kg')
#print('\nSolution 2 :', sol2,'kg')
#print('\nSolution 3 :', sol3,'kg \n \n')
sol3sunxs = sol1xs/Msun
print("La masse du compagnon pour l'excès IR est de %.2f M_sol" %sol3sunxs)
```
